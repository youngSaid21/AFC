## Importation des packages et des données

```python
import pandas as pd
from fanalysis.ca import CA
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Chargement des données
df = pd.read_excel("Nobel_avecMath.xlsx")
df
```

## Reconstruction du tableau de données

```python
# Extraire la colonne 'Nobel_avecMath (2)' et la stocker dans la variable 'country'
country = df['Nobel_avecMath (2)']

# Récupérer tous les noms de colonnes sauf la première
col_names = list(df.columns)[1:]

# Extraire les valeurs des colonnes sélectionnées sous forme de tableau NumPy
X = df[col_names].values

# Créer un DataFrame vide
df_result = pd.DataFrame({})

# Ajouter une colonne 'Country' dans le DataFrame 'df_result' avec les valeurs de 'country', en sautant la première ligne
df_result['Country'] = country[1:]

# Nombre de colonnes dans X (correspondant au nombre de caractéristiques/variables)
n = len(X[0])

# Boucle pour ajouter chaque colonne de X (à partir de la seconde ligne) dans df_result
for i in range(n):
    # Ajouter une colonne avec les valeurs correspondantes à chaque variable
    # 'X[0][i]' correspond au nom de la variable, 'X[1:, i]' contient les valeurs
    df_result[X[0][i]] = X[1:, i]
```

```python
df_result
```

```python
# Reindexer le dataframe
df_result.reset_index(drop=True, inplace=True)
```

```python
df_result
```

## Description des données

```python
# Extraction des noms de colonnes et de la colonne "Departement"
col_names = list(df_result.columns)[1:]
country_names = df_result['Country']
```

```python
# Extraction des données numériques
numeric_data = df_result[col_names].values
```

### Distribution des prix nobels des pays par discipline

```python
# Calcul du profil ligne
row_sum = np.sum(numeric_data, axis=1)
row_profile = numeric_data / row_sum[:, np.newaxis]

# Affichage
print("Profil ligne:")
print(pd.DataFrame(row_profile, index=country_names, columns=col_names))
```

### Distribution des nobels obtenus par les discipline selon le pays

```python
# Calcul du profil colonne
col_sum = np.sum(numeric_data, axis=0)
col_profile = numeric_data / col_sum[np.newaxis, :]

# Affichage
print("Profil colonne:")
print(pd.DataFrame(col_profile, index=country_names, columns=col_names))
```

## Analyse des prix nobels

```python
# Extraction des noms de colonnes
col_names = list(df_result.columns)[1:]

# Calcul de la somme des prix nobel pour chaque discipline
total_nobel = df_result[col_names].sum()

# Calcul des proportions pour chaque discipline
prop_nobel = total_nobel 
# Tri des proportions de la plus grande à la plus petite
prop_nobel = prop_nobel.sort_values(ascending=False)

# Création du diagramme en barres
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(prop_nobel.index, prop_nobel)

# Ajout de labels et de titres
ax.set_xlabel('Discipline')
ax.set_ylabel('Proportion des prix nobels')
ax.set_title('Répartition des prix nobels par discipline (par ordre décroissant)')
# Rotation des noms des candidats
plt.xticks(rotation=45, ha='right')

# Affichage du diagramme
plt.show()

prop_nobel
```

```python
# Extraction des noms de colonnes
col_names = list(df_result.columns)[1:]

# Calcul de la somme des votes pour chaque candidat
total_nobel = df_result[col_names].sum()

# Calcul des proportions pour chaque candidat
prop_nobel = total_nobel / total_nobel.sum()

# Tri des proportions de la plus grande à la plus petite
prop_nobel = prop_nobel.sort_values(ascending=False)

# Création du diagramme en barres
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(prop_nobel.index, prop_nobel)

# Ajout de labels et de titres
ax.set_xlabel('Discipline')
ax.set_ylabel('Proportion des prix nobels')
ax.set_title('Répartition des prix nobles par discipline (par ordre décroissant)')
# Rotation des noms des candidats
plt.xticks(rotation=45, ha='right')

# Affichage du diagramme
plt.show()

prop_nobel
```

## Analyse factorielle des composantes (AFC)

### Instaciation de l'AFC avec FAnalysis

```python
# Redéfinir l'index sur Country
df = df_result.set_index("Country")
df
```

```python
# Conversion de type de données
df = df.apply(pd.to_numeric, errors='coerce')
# Instanciation de la classe CA
new_ca = CA(row_labels=df.index.values,col_labels=df.columns.values,stats=True)
```

```python
print(df.dtypes)
```

```python
# Entrainment de l'AFC
new_ca.fit(df.values)
```

### Choix du nombre d'axe factoriel

```python
# Identification des valeurs propres
new_ca.eig_
```

```python
# Graphique des valeurs propres
new_ca.plot_eigenvalues()
```

```python
# Graphique des valeurs propres en pourcentage
new_ca.plot_eigenvalues(type="percentage")
```

```python
# Graphique cummulative des valeurs propres en pourcentage
new_ca.plot_eigenvalues(type="cumulative")
```

### Analyse des qualités de représentation (cos 2)

#### Analyse suivant le profil ligne (Country)

```python
# Classement des points colonnes en fonction de leur contribution sur les deux premiers axes
print("Premier axe factoriel")
new_ca.plot_row_cos2(num_axis=1)
print("Second axe factoriel")
new_ca.plot_row_cos2(num_axis=2)
```

#### Analyse suivant le profil colonne (Discipline)

```python
print("Premier axe factoriel")
new_ca.plot_col_cos2(num_axis=1)
print("Second axe factoriel")
new_ca.plot_col_cos2(num_axis=2)
```

### Analyses des contributions

#### Profil lignes (Country)

```python
print("Premier axe")
new_ca.plot_row_contrib(num_axis=1)
print("Second axe")
new_ca.plot_row_contrib(num_axis=2)
```

#### Profil colonne (Disciplines)

```python
print("Premier axe")
new_ca.plot_col_contrib(num_axis=1)
print("Second axe")
new_ca.plot_col_contrib(num_axis=2)
```
